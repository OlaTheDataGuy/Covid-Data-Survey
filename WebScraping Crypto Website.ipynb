### Use BeautifulSoup to scrape Crypto Website

```python
### Import Necessary libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd
import sqlalchemy
```

```python
## Store website link

url = 'https://www.coingecko.com/en'
```

```python
###  Get requests from website link


response = requests.get(url)
```

```python
### Use BeautifulSoup to get access to content & HTML 


soup = BeautifulSoup(response.content, 'html.parser')
```

```python
### Status Code


response.status_code
```

```python
### View soup crosscheck with website

soup
```

```python
### store all results for first page in variable


res = soup.find('table', {'class':'table-scrollable'}).find('tbody').find_all('tr')
```

```python
###  Use For-Loop to extract needed data


name = []
price = []
change_1h = []
change_24h = []
change_7d = []
volume_24h = []
market_cap=[]

for result in res:
    
    # name
    try:
        name.append(result.find('a', {'class':'tw-hidden lg:tw-flex font-bold tw-items-center tw-justify-between'}).get_text().strip()) 
    except:
        name.append('n/a')
    
    # price
    try:
        price.append(result.find('span', {'class':'no-wrap'}).get_text().strip())
    except:
        price.append('n/a')
    
    # change_1h
    try:
        change_1h.append(result.find('td', {'class':'td-change1h'}).get_text().strip())
    except:
        change_1h.append('n/a')
        
    # change_24h
    try:
        change_24h.append(result.find('td', {'class':'td-change24h'}).get_text().strip())
    except:
        change_24h.append('n/a')
    
    # change_7d
    try:
        change_7d.append(result.find('td', {'class':'td-change7d'}).get_text().strip())
    except:
        change_7d.append('n/a')
    
    # volume_24h
    try:
        volume_24h.append(result.find('td', {'class':'td-liquidity_score'}).get_text().strip())
    except:
        volume_24h.append('n/a')
    
    # market_cap
    try:
        market_cap.append(result.find('td', {'class':'td-market_cap'}).get_text().strip())
    except:
        market_cap.append('n/a')


```

```python
### Create Pandas Dataframe for first page


df = pd.DataFrame({'Name': name, 'Price':price, '1hr %Change':change_1h, '24hr %Change': change_24h,
                       '7d %Change': change_7d, '24hr Volume': volume_24h, 'Market Cap': market_cap})
```

```python
####  Scrape multiple Pages - Pagination


name = []
price = []
change_1h = []
change_24h = []
change_7d = []
volume_24h = []
market_cap=[]

for i in range(1,11):
    print(i)
    
    #website 
    
    url ='https://www.coingecko.com/en?page='+str(i)

    
    # request to website link
    
    response = requests.get(url)
    
    
    # soup 
    
    soup = BeautifulSoup(response.content, 'html.parser')
    
    
    # store results
    
    res = soup.find('table', {'class':'table-scrollable'}).find('tbody').find_all('tr')
    
    
    for result in res:
    
        # name
        try:
            name.append(result.find('a', {'class':'tw-hidden lg:tw-flex font-bold tw-items-center tw-justify-between'}).get_text().strip()) 
        except:
            name.append('n/a')

        # price
        try:
            price.append(result.find('span', {'class':'no-wrap'}).get_text().strip())
        except:
            price.append('n/a')

        # change_1h
        try:
            change_1h.append(result.find('td', {'class':'td-change1h'}).get_text().strip())
        except:
            change_1h.append('n/a')

        # change_24h
        try:
            change_24h.append(result.find('td', {'class':'td-change24h'}).get_text().strip())
        except:
            change_24h.append('n/a')

        # change_7d
        try:
            change_7d.append(result.find('td', {'class':'td-change7d'}).get_text().strip())
        except:
            change_7d.append('n/a')

        # volume_24h
        try:
            volume_24h.append(result.find('td', {'class':'td-liquidity_score'}).get_text().strip())
        except:
            volume_24h.append('n/a')

        # market_cap
        try:
            market_cap.append(result.find('td', {'class':'td-market_cap'}).get_text().strip())
        except:
            market_cap.append('n/a')
```

```python
### Create Pandas Dataframe for all pages


df = pd.DataFrame({'Name': name, 'Price':price, '1hr %Change':change_1h, '24hr %Change': change_24h,
                       '7d %Change': change_7d, '24hr Volume': volume_24h, 'Market Cap': market_cap})
```
